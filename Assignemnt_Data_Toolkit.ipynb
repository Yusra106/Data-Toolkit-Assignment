Ques2. How does broadcasting work in NumPy ?

In NumPy, broadcasting is a mechanism that enables arithmetic operations on arrays with different shapes, as if the smaller array were "stretched" or "broadcast" to match the shape of the larger array. This avoids the need for explicit looping and reshaping, leading to more efficient and concise code.


Here's how broadcasting works:
1. Shape Compatibility:
Broadcasting operates under specific rules to determine if two arrays can be aligned for operations. If the arrays have different numbers of dimensions, the smaller array is padded with ones on the left.

2. Dimension-wise Matching:
Two dimensions are compatible if they have the same size, or if one of them has a size of 1.

3. Implicit Stretching:
If the dimensions are compatible, the smaller array is effectively "stretched" or "broadcast" across the larger array so that they have compatible shapes. This means that the smaller array acts as if it's repeated or copied along the dimension where it has a size of 1.

4. Operation Execution:
Once the shapes are compatible, the arithmetic operation is then performed element-wise.

Ques8. What does the term “vectorized operation” mean in NumPy ?

In NumPy, a vectorized operation refers to performing operations on entire arrays (vectors, matrices, etc.) without using explicit loops (like for or while loops). Instead of iterating over elements one by one, NumPy applies operations element-wise under the hood, which is much faster and more efficient due to underlying optimizations.

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Vectorized addition
c = a + b
print(c)  # Output: [5 7 9]
```

This operation is vectorized. You didn't write a loop, but each element of a was added to the corresponding element of b

NumPy handles multidimensional arrays using the powerful ndarray (N-dimensional array) object. Here’s a breakdown of how it works and how it manages multidimensional data:

1. The ndarray Object
At the core of NumPy is the ndarray class, which supports arbitrary dimensions (1D, 2D, 3D, and beyond).

You can think of a 2D array as a matrix, 3D as a cube, and higher dimensions as tensors.

```python
import numpy as np

# 1D array
a = np.array([1, 2, 3])

# 2D array
b = np.array([[1, 2, 3], [4, 5, 6]])

# 3D array
c = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
```

.shape: Tuple representing the size of the array in each dimension.

.ndim: Number of dimensions.

```python
print(c.shape)  # (2, 2, 2)
print(c.ndim)   # 3
```

3. Memory Layout (C-style / Fortran-style)
NumPy stores arrays in row-major (C-style) order by default.

You can control this with the order parameter (e.g., order='F' for Fortran-style column-major).

This layout is used to optimize performance, especially for vectorized operations and broadcasting.

4. Broadcasting
NumPy can perform operations on arrays of different shapes, automatically "stretching" them when compatible.

Ques17. Explain the difference between apply() and map() in Pandas ?

Both apply() and map() are used in Pandas to apply functions to data, but they are used in different contexts and have different scopes.

🔹** map()**
Used with: Series only (not DataFrames).

Purpose: Applies a function element-wise to each value in the Series.

Common use case: Simple value transformation or substitution in a single column.

```python
import pandas as pd

s = pd.Series([1, 2, 3])
s.map(lambda x: x * 10)
```

** apply()**
Used with: Both Series and DataFrames.

Purpose:

On Series: behaves similarly to map().

On DataFrame: allows you to apply a function along rows or columns using axis argument (axis=0 for columns, axis=1 for rows).

Common use case: More flexible operations that involve row/column logic or applying functions that return more than one value.

```python
s.apply(lambda x: x * 10)
```

Ques21. Why is NumPy’s array slicing faster than Python’s list slicing ?

NumPy’s array slicing is faster than Python’s list slicing due to several key reasons related to memory layout, data type homogeneity, and low-level optimizations. Here's a breakdown:

1. **Contiguous Memory Storage**
NumPy arrays are stored in contiguous blocks of memory (like in C).

Python lists are arrays of pointers to separate Python objects, which may be scattered in memory.

This allows NumPy to take advantage of vectorized operations and low-level memory copying, which are much faster.

2. **Homogeneous Data Types**
NumPy arrays contain elements of a single data type (e.g., float64, int32).

Python lists can contain mixed types (ints, strings, other lists, etc.), so Python has to check the type of each element during slicing.

NumPy skips type checks and uses fixed-size data chunks, making slicing a lightweight pointer operation.

3. **View vs Copy**
NumPy slicing usually returns a view — it doesn't copy data, just creates a new "window" into the same memory block.

Python list slicing always returns a copy — it has to allocate new memory and individually reference each element.

NumPy avoids the overhead of copying unless you explicitly request it.

 4. **C Implementation**
NumPy is implemented in C, and uses highly optimized loops for slicing and indexing.

Python lists are high-level objects managed by the Python interpreter, which is slower.

 NumPy leverages compiled code and SIMD (Single Instruction, Multiple Data) instructions when possible.

8. Create a line plot with multiple lines using Matplotlib .

```python
import matplotlib.pyplot as plt

# Sample data
x = [0, 1, 2, 3, 4, 5]
y1 = [0, 1, 4, 9, 16, 25]
y2 = [0, 1, 2, 3, 4, 5]
y3 = [0, -1, -2, -3, -4, -5]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='y = x^2', color='blue', marker='o')
plt.plot(x, y2, label='y = x', color='green', linestyle='--')
plt.plot(x, y3, label='y = -x', color='red', linestyle='-.')

# Add labels and title
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Multiple Line Plot Example')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold .

```python
import pandas as pd

# Step 1: Create a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Score': [85, 42, 90, 67, 77]
}

df = pd.DataFrame(data)

# Step 2: Filter rows where 'Score' is greater than a threshold (e.g., 70)
threshold = 70
filtered_df = df[df['Score'] > threshold]

print(filtered_df)
```

10. Create a histogram using Seaborn to visualize a distribution .

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a histogram of the 'total_bill' column
sns.histplot(data=tips, x='total_bill', bins=20, kde=True)

# Add title and labels
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")

# Show the plot
plt.show()
```

11.Perform matrix multiplication using NumPy .

```python
import numpy as np

# Define two matrices
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Matrix multiplication using @ operator (preferred)
result = A @ B

# Alternatively, use np.dot(A, B)
# result = np.dot(A, B)

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

print("\nResult of A & B:")
print(result)
```

13. Create a 3D scatter plot using Plotly.

```python
import plotly.graph_objects as go

# Sample data (replace with your actual data)
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 5, 3]
z = [3, 1, 5, 2, 4]

# Create the 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z, mode='markers')])

# Customize the plot (optional)
fig.update_layout(title='3D Scatter Plot',
                  scene=dict(xaxis_title='X Axis',
                             yaxis_title='Y Axis',
                             zaxis_title='Z Axis'))

# Show the plot
fig.show()
```
